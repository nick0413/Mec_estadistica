```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


data = [np.random.random((10, 10)) for _ in range(100)]

fig, ax = plt.subplots()
cax = ax.imshow(data[0], interpolation='nearest')
fig.colorbar(cax)

def init():
    cax.set_data(data[0])
    ax.set_xticks([])
    return [cax]

def update(i):
    cax.set_data(data[i])
    cax.set_cmap('inferno')
    cax.set_clim(0,4)
    ax.set_xticks([])
    return [cax]

ani = animation.FuncAnimation(fig, update, frames=range(len(data)), init_func=init, blit=True)
ani.save('animation5.gif', writer='imagemagick', fps=10)
plt.show()
```
